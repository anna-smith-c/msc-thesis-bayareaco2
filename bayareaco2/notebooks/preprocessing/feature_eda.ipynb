# **Exploratory Data Analysis: Extracted Feature Data**

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
path = Path.cwd().parent.parent.parent / 'Data' / 'features.csv'
df = pd.read_csv(path)
```

## Land Use

```python
lu_50m = ['Built_Area_area_50m', 'Rangeland_area_50m', 'Trees_area_50m', 'Water_area_50m', 'Bare_Ground_area_50m', 'Crops_area_50m', 'Flooded_Vegetation_area_50m']
```

```python
def pie_lu(distance):
    # List of land use types
    land_use_types = [
        'Built_Area_area', 
        'Rangeland_area', 
        'Trees_area', 
        'Water_area', 
        'Bare_Ground_area', 
        'Crops_area', 
        'Flooded_Vegetation_area'
    ]
    
    # Generate the array with the correct distance labels
    lu_array = [f'{land_use}_{distance}m' for land_use in land_use_types]
    
    lu_mean = df[lu_array].mean()
    lu_values = lu_mean[lu_mean > 0].values
    lu_labels = lu_mean[lu_mean > 0].index

    wedges, texts = plt.pie(
    lu_values # This adds the percentage labels to the pie chart
)

    plt.legend(
        wedges, lu_labels, title='Land Use Categories', loc="center left", 
        bbox_to_anchor=(1, 0, 0.5, 1)
    )
    plt.title(f'Land Use {distance}m')
```

```python
buffers = [50, 100, 200, 300, 500, 1000, 2000, 3000, 4000, 5000]

fig, axes = plt.subplots(nrows=5, ncols=2, figsize=(14, 20))
axes = axes.flatten()  # Flatten the 2D array to 1D for easy iteration

for i, buffer in enumerate(buffers):
    plt.sca(axes[i])  # Set the current axis
    pie_lu(buffer)
    
plt.tight_layout()
plt.show()
```
